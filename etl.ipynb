```python
import pandas as pd
```

```python
df = pd.read_csv('personal_income.csv')
df.head(11)
```

```python
df.dtypes
```

```python
# Assuming 'df' is your DataFrame
columns_to_clean = ['2020', '2021', '2022']
for column in columns_to_clean:
    df[column] = df[column].apply(lambda x: x.split(' - ') if isinstance(x, str) and '-' in x else [x])
    #df[column] = df[column].apply(lambda x: [y.strip('% ') for y in x] if isinstance(x, list) else x)
```

```python
df.head(20)
```

```python
df[['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']] = df[['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']].replace({',': '', '%': ''}, regex=True).astype(float)
df.head(11)
```
