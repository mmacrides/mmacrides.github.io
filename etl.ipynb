```python
import pandas as pd
```

```python
df = pd.read_csv('personal_income.csv')
df.head(11)
```

```python
# Find averaged data from ranged data
columns_to_clean = ['2020', '2021', '2022']
for column in columns_to_clean:
    df[column] = df[column].replace({',': '', '%': ''}, regex=True)
    df[column] = df[column].apply(lambda x: x.split(' - ') if isinstance(x, str) and '-' in x else [x])
    df[column] = df[column].apply(lambda x: round(sum(float(y) for y in x) / len(x), 1) if len(x) == 2 else round(float(x[0]), 1) if len(x) == 1 else x)
#df.loc[df['LineCode'].isin([6, 7]), columns_to_clean] = df.loc[df['LineCode'].isin([6, 7]), columns_to_clean].astype(str) + '%'
# Clean the rest
df[['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']] = df[['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']].replace({',': '', '%': ''}, regex=True).astype(float)
df.head(11)
```

```python
df.to_csv('personal_income_formatted.csv', index=False)
```
